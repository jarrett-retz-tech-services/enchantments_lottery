## Lottery Information

Applicants were allowed to choose three different entry dates for their trip, group sizes, and permit zone. For example:

-   7/25/2021, Core Enchantment Zone, 4 people
-   8/16/2021, Core Enchantment Zone, 2 people
-   9/2/2021, Stuart Lake Zone, 8 people

However, they didn't have to choose three options. Some applicants only chose to submit one option. Just looking at the `Preferred Entry Date 1`, applicants applied for trips between May 15 and October 30. Applicants chose group sizes between 1-8.

A total of 36,695 applications made up the lottery in 2021 (the website PDF says 39,695 but the data spreadsheet only has 36,695 rows, I believe the website PDF is a typo).

### How Lottery/Quota Works

Per [recreation.gov](https://www.recreation.gov/permits/233273):

> A group may consist of up to eight people, must camp in the Zone indicated on their permit for up to 14 days. The number of overnight groups/people that may enter each day is split by zone during permit season:

-   \*Core Enchantments Zone : 24 people per day (cannot increase group size after lottery)
-   Colchuck Lake Zone: three groups per day
-   Stuart Lake Zone: four groups per day
-   Snow Lakes Zone: five groups per day
-   Eightmile/Caroline Zone: three groups per day

### Awarded Permits

Just because a permit is awarded, doesn't mean that the applicant went on the trip. Looking at the `Results Status` there are possible values of:

-   Unsuccessful
-   Accepted
-   Cancelled
-   No Response
-   Declined

**In 2021, 352 applications didn't response to their awarded lottery, which they had won!** This statistic is even more shocking because I applied for a lottery in 2021 and was `Unsuccessful`.

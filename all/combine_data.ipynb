```python
import pandas as pd
import numpy as np

# Import the 2023 enchantments data
df_2023 = pd.read_csv(
    '2023_results_cleaned.csv',    
    # Import was failing to parse date columns, so I
    # to add the column names
    parse_dates=[
        "preferred_entry_date_1",
        "preferred_entry_date_2",
        "preferred_entry_date_3",
        "awarded_entry_date",
    ],
    date_format="%m-%d-%Y",  # Align format with export format
    na_filter=False,  # Do not convert 'N/A' to NaN
)

# Add a column to indicate the year
df_2023["year"] = 2023

df_2023.head()
```

```python
# Import the 2022 enchantments data
df_2022 = pd.read_csv(
    '2022_results_cleaned.csv',    
    # Import was failing to parse date columns, so I
    # to add the column names
    parse_dates=[
        "preferred_entry_date_1",
        "preferred_entry_date_2",
        "preferred_entry_date_3",
        "awarded_entry_date",
    ],
    date_format="%m-%d-%Y",  # Align format with export format
    na_filter=False,  # Do not convert 'N/A' to NaN
)

# Add a column to indicate the year
df_2022["year"] = 2022

# Change the column awared_entrance to awarded_entrance_code_name
df_2022 = df_2022.rename(columns={'awarded_entrance': 'awarded_entrance_code_name'})

df_2022.head()
```

```python
# Import the 2021 enchantments data
df_2021 = pd.read_csv(
    '2021_results_cleaned.csv',    
    # Import was failing to parse date columns, so I
    # to add the column names
    parse_dates=[
        "preferred_entry_date_1",
        "preferred_entry_date_2",
        "preferred_entry_date_3",
        "awarded_entry_date",
    ],
    date_format="%m-%d-%Y",  # Align format with export format
    na_filter=False,  # Do not convert 'N/A' to NaN
)

# Change the columns 'preferred_division_1' to 'preferred_zone_1' and 
# 'preferred_division_2' to 'preferred_zone_2' and 'preferred_division_3' to 'preferred_zone_3'
df_2021 = df_2022.rename(columns={
    'preferred_division_1': 'preferred_zone_1',
    'preferred_division_2': 'preferred_zone_2',
    'preferred_division_3': 'preferred_zone_3'
})

# Add a column to indicate the year
df_2021["year"] = 2021

df_2021.head()
```

```python
# Import 2020 enchantments data
df_2020 = pd.read_csv(
    '2020_results_cleaned.csv',    
    # Import was failing to parse date columns, so I
    # to add the column names
    parse_dates=[
        "preferred_entry_date_1",
        "preferred_entry_date_2",
        "preferred_entry_date_3",
        "awarded_entry_date",
    ],
    date_format="%m-%d-%Y",  # Align format with export format
    na_filter=False,  # Do not convert 'N/A' to NaN
)

# Add a column to indicate the year
df_2020['year'] = 2020

df_2020.head()
```

```python
# Create a list of all the columns in the dataframes
columns_23 = df_2023.columns.tolist()
columns_22 = df_2022.columns.tolist()
columns_21 = df_2021.columns.tolist()
columns_20 = df_2020.columns.tolist()

# Combine all the column lists
all_columns = columns_23 + columns_22 + columns_21 + columns_20

# Count the number of times each column appears in the list
column_counts = {x: all_columns.count(x) for x in all_columns}

column_counts
```

```python
# Combine all the dataframes
df_all = pd.concat([df_2020, df_2021, df_2022, df_2023], ignore_index=True)

df_all
```

```python
# Fill in NaN values for minimum_acceptable_group_size_1, minimum_acceptable_group_size_2, and minimum_acceptable_group_size_3
df_all['minimum_acceptable_group_size_1'] = df_all['minimum_acceptable_group_size_1'].fillna(0)
df_all['minimum_acceptable_group_size_2'] = df_all['minimum_acceptable_group_size_2'].fillna(0)
df_all['minimum_acceptable_group_size_3'] = df_all['minimum_acceptable_group_size_3'].fillna(0)

# Save the combined dataframe to a csv file
df_all.to_csv('combined_results.csv', index=False, date_format="%m-%d-%Y")
```

```python
# It may be better to break up each individual entry into its own row, so that the data can be analyzed more easily.
preferred_options = [1, 2, 3]

# Columns that every dataframe will have
shared_columns = [
    "results_status",
    "awarded_preference",
    "awarded_entry_date",
    "awarded_entrance_code_name",
    "awarded_group_size",
    "processing_sequence",
    "state",
    "year",
]
new_dataframes = []

# Iterate over each option number creating a new dataframe for each
for option in preferred_options:
    # Get the columns for the current option
    columns = [
        f"preferred_zone_{option}",
        f"preferred_entry_date_{option}",
        f"minimum_acceptable_group_size_{option}",
    ]
    # Create a new dataframe for the current option
    df_option = df_all[columns + shared_columns].copy()
    # Rename the columns to remove the option number
    df_option.columns = [
        "preferred_zone",
        "preferred_entry_date",
        "minimum_acceptable_group_size",
    ] + shared_columns
    # Add a column to indicate if the permit was awarded for the current option
    df_option["awarded"] = df_option["awarded_preference"] == option
    df_option["preferred_option"] = option

    # Append the new dataframe to the list of dataframes
    new_dataframes.append(df_option)

# Concatenate the list of dataframes into a single dataframe
df_split = pd.concat(new_dataframes)

# Drop rows where the preferred division is N/A
df_split = df_split[df_split["preferred_zone"] != "N/A"]

# Check the new dataframe
df_split
```

```python
# Add the month of the preferred entry date to the dataframe
import calendar

# Get the month as an integer
df_split["preferred_entry_date" + "_month"] = df_split["preferred_entry_date"].dt.month
# Get the month as a string
df_split["preferred_entry_date" + "_month"] = df_split[
    "preferred_entry_date" + "_month"
].apply(lambda x: calendar.month_name[x])

# Add the day of the week columns based on preferred entry date
df_split["preferred_entry_date" + "_day"] = df_split[
    "preferred_entry_date"
].dt.day_name()

df_split
```

```python
# Save the combined split dataframe to a csv file
df_split.to_csv('combined_results_split.csv', index=False, date_format="%m-%d-%Y")
```

```python
# Create a datatframe removing the entries that never had a chance to win
# because the applicatant had already been awarded a permit

# Find where the preferred option was greater than or equal to the awarded preference
preferred_option_less_than_awarded_option = (
    df_split["preferred_option"] >= df_split["awarded_preference"]
)

df_split_actual = df_split[
    preferred_option_less_than_awarded_option
].copy()

df_split_actual
```
